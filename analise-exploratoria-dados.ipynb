```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importando os dados
dados = pd.read_csv('dados\dataset-store.csv')
```

```python
# Verificar se há dados ausentes em todo o DataFrame
print(dados.isnull().sum())
```

```python
# Remover linhas com dados ausentes
dados = dados.dropna()
```

```python
# Resumo estatístico para variáveis numéricas
resumo_numericas = dados.describe()
print(resumo_numericas)
```

```python
# Histograma para uma variável numérica
dados['Quantidade'].hist()
plt.title('Distribuição de quantidade de Produtos por Venda')
plt.show()
```

```python
# Contagem de valores únicos para uma variável categórica
print(dados['Categoria'].value_counts())

# Gráfico de barras para uma variável categórica
sns.countplot(x='Categoria', data=dados)
plt.xticks(rotation=90)
plt.title('Distribuição de Categorias')
plt.show()
```

```python
# Matriz de correlação
corr = dados.corr()
print(corr)

# Mapa de calor de correlação
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlação')
plt.show()
```

```python
# Gráfico de dispersão entre duas variáveis numéricas
sns.scatterplot(x='Quantidade', y='Valor_Venda', data=dados)
plt.title('Valor de Venda x Quantidade')
plt.show()
```
